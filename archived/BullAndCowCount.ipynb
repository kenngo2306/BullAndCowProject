```
result = "TOAN"
guess = "NOAR"

r_index = 0
g_index = 0

bull = 0
cow = 0

while (r_index <= 3):
  g_index = 0
  while (g_index <= 3):
    #print(result[r_index] + guess[g_index])
    if(r_index == g_index and result[r_index] == guess[g_index]):
      bull += 1
      break
    elif (result[r_index] == guess[g_index]):
      cow += 1
      break
    g_index += 1
  r_index += 1
  
print ('bull=' + str(bull))
print ('cow=' + str(cow))
```
